```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('myothermoviedb.csv', engine='python')
```

```python
data[['Vote_Average','Vote_Count']] = data[['Vote_Average','Vote_Count']].astype(float)
```

```python
data.tail(11)
```

```python
data.describe()
```

```python
x=data["Vote_Average"]
y=data["Popularity"]
plt.scatter(x,y)
plt.title("Average Vote and Popularity")
plt.xlabel("Vote Average")
plt.ylabel("Popularity")
```

```python
sns.regplot(x="Vote_Count", y="Vote_Average", data=data)
plt.ylim(0,)
```
